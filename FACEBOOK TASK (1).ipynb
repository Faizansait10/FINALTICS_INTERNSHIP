### Question 1: How does the time of upload (status_published) affect num_reaction?

```python
df.head()
```

```python
df['status_published']=pd.to_datetime(df['status_published'])
df.head()
```

```python
# Extract the hour of the day the post was published
df['publish_hour'] = df['status_published'].dt.hour #.dt is accessor to access the datetime
#specific functionalities
df['publish_hour'].head()
df.head()
```

```python
# Extract the day of the week (Monday=0, Tuesday=1, ..., Sunday=6)
df['publish_day_of_week'] = df['status_published'].dt.dayofweek
df['publish_day_of_week'].head()
df.head()
```

```python
#extract the month of the year the post was pulished
df['publish_month']=df['status_published'].dt.month
df.head() #1-->jan 2-->feb ....12-->decemebr
```

```python
# Let's see the first few rows of the DataFrame with the new features
print("\nDataFrame with extracted time features:")
df[['status_published', 'publish_hour', 'publish_day_of_week', 'publish_month']].head()
```

```python
columns_drop=['Column1','Column2','Column3','Column4']
df=df.drop(columns=columns_drop)
```

```python
df[['status_published', 'publish_hour', 'publish_day_of_week', 'publish_month','num_reactions']].head()
```

#### Why extract these specific features?

##### Hour of the Day: User activity on social media platforms often varies throughout the day. We want to see if posts published at certain times (e.g., peak usage hours) tend to receive more reactions.
##### Day of the Week: Engagement might also differ based on the day of the week. For example, weekend posts might have different interaction patterns than weekday posts.
##### Month of the Year: There could be seasonal trends or specific events during certain months that influence engagement.

```python
from sklearn.cluster import KMeans

# Initialize the K-Means model with a chosen number of clusters (let's start with 3)
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10) # Added n_init for future warning

# Fit the K-Means model to the scaled data
kmeans.fit(cluster_data_scaled)

# Get the cluster labels assigned to each data point
cluster_labels = kmeans.labels_

# Add the cluster labels back to the original DataFrame for analysis
df['cluster'] = cluster_labels

# Display the count of data points in each cluster
print(f"\nDistribution of data points across {n_clusters} clusters:")
print(df['cluster'].value_counts())

# Display the cluster centers (in the scaled feature space)
print("\nCluster centers (scaled data):")
print(kmeans.cluster_centers_)
```

### Question 4: Use the elbow method to find the optimum number of clusters.

```python
range_of_clusters = range(1, 11)
wcss = []

for i in range_of_clusters:
    # Initialize and fit the K-Means model for the current number of clusters
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(cluster_data_scaled)

    # Append the inertia (WCSS) to our list
    wcss.append(kmeans.inertia_)

# Print the WCSS values for each number of clusters
print("\nWCSS values for different numbers of clusters:")
print(wcss)
```

```python
# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range_of_clusters, wcss, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squared Errors (WCSS)')
plt.grid(True)
plt.xticks(range_of_clusters)
plt.show()

#The elbow method suggests that the optimal number of clusters for this dataset is likely 3 or 4. 
#Choosing either of these numbers would represent a good balance between minimizing the 
#within-cluster variance and avoiding overfitting by having too many clusters.
```
